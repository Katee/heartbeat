```python
import subprocess as sp
import numpy as np
import random

FFMPEG = "ffmpeg"

videofile = "heartbeat.mp4"
video_width = 1280
video_height = 720

# Graphing helper function
def setup_graph(title='', x_label='', y_label='', fig_size=[16, 8]):
    fig = plt.figure()
    if fig_size != None:
        fig.set_size_inches(fig_size[0], fig_size[1])

    ax = fig.add_subplot(111)
    ax.set_title(title)
    
    ax.set_xlabel(x_label)
    ax.set_ylabel(y_label)
```

```python
def create_sampling_function(video_width, video_height, num_samples=300):
    """ Returns a function that returns the average amount of red in a frame """
    samples = [[random.randint(0, video_height - 1), random.randint(0, video_width - 1)] for i in xrange(0, num_samples)]
    
    def sample_frame(frame):
        """ Returns the average amount of red in a frame by randomly sampling some pixels """
        reds = [frame[sample[0]][sample[1]][0] for sample in samples]
        return np.average(reds)
    
    return sample_frame
```

```python
def video_frames(videofile, video_width, video_height, frames=1):
    pipe = sp.Popen([ FFMPEG,
                   "-i", videofile,
                   "-f", "image2pipe",
                   "-vframes", str(frames),
                   "-pix_fmt", "rgb24",
                   "-vcodec", "rawvideo", "-"],
                   stdout = sp.PIPE)
    
    current_frame = 0
    while current_frame < frames:
        # one frame is video_width * vieo_height * 3 bytes
        raw_image = pipe.stdout.read(video_width*video_height*3)
        image = np.fromstring(raw_image, dtype='uint8').reshape((video_height,video_width,3))
        current_frame += 1
        yield image
```

```python
import re
from decimal import Decimal

def get_video_duration(videofile):
    process = sp.Popen([FFMPEG, '-i', videofile], stdout=sp.PIPE, stderr=sp.STDOUT)
    stdout, stderr = process.communicate()
    
    duration = re.search(r"Duration:\s{1}(?P<hours>\d+?):(?P<minutes>\d+?):(?P<seconds>\d+\.\d+?),", stdout, re.DOTALL).groupdict()
    duration = {k: Decimal(v) for k, v in duration.iteritems()}
    seconds = (60 * 60 * duration['hours']) + (60 * duration['minutes']) + duration['seconds']
    
    return seconds

def get_video_fps(videofile):
    process = sp.Popen([FFMPEG, '-i', videofile], stdout=sp.PIPE, stderr=sp.STDOUT)
    stdout, stderr = process.communicate()
    
    # this will break if fps shows up elsewhere in the info string
    return Decimal(re.search(r"(?P<fps>\d+\.?\d+?) fps,", stdout, re.DOTALL).groupdict()['fps'])
```

```python
video_duration = get_video_duration(videofile)
video_fps = get_video_fps(videofile)
num_frames = int(video_duration * video_fps) - 1

frames = video_frames(videofile, video_width, video_height, num_frames)
sampler = create_sampling_function(video_width, video_height)

reds = [sampler(frames.next()) for x in xrange(0, num_frames)]
```

```python
setup_graph(x_label='Time (in Frames)', y_label='Amount of Red')
plot(reds)
```

```python
deriv = []

for i in xrange(0, len(reds) - 1):
    v = np.average([reds[i] / reds[i + 1]])
    deriv.append(v)
    
def strech_and_operate(data, window, operation):
    half_window = window / 2
    data_stretched = (([data[0]] * window) + data + ([data[-1]] * window))[half_window:]
    new_data = []
    
    for i in xrange(half_window, len(data_stretched) - window + half_window):
        new_data.append(operation(data_stretched[i - half_window:i + window - half_window]))
        
    return new_data

def get_moving_std(data, window=80):
    def std(d):
        return 1 + np.std(d) * 1.5
    return strech_and_operate(data, window, std)

def get_moving_mean(data, window=80):
    return strech_and_operate(data, window, np.mean)

moving_std = get_moving_std(deriv, 90)
moving_mean = get_moving_mean(deriv, 30)
```

```python
setup_graph(x_label='Time (in Frames)', y_label='Rate of Change in Red')

deriv2 = [np.abs(deriv[i] / moving_mean[i] - 1) + 1 for i in xrange(0, len(deriv))]
deriv2 = [(deriv[i] / moving_mean[i]) for i in xrange(0, len(deriv))]

plot(deriv2, 'grey')

plot(moving_std)
```

```python
def get_top_frames(frames, crit_values, min_frames_between_heartbeat=10):
    top_frames = []
    
    max_frame = None
    max_frame_value = 0
    
    for i in xrange(0, len(frames)):
        p = deriv2[i]
        crit_value = crit_values[i]
        if p > crit_value and p > max_frame_value:
            max_frame = i
            max_frame_value = p
        elif p < crit_value and max_frame != None and i - max_frame > min_frames_between_heartbeat:
            top_frames.append(max_frame)
            max_frame = None
            max_frame_value = 0
    return top_frames[1:] # have to throw away the first one as it might only be a partial heartbeat
```

```python
top_frames = get_top_frames(deriv2, moving_std)
print top_frames
```

```python
frames_between_beats = []

for i in xrange(0, len(top_frames) - 1):
    frames_between_beats.append(top_frames[i + 1] - top_frames[i])
    
print frames_between_beats # number of frames between each heartbeat
```

```python
def smooth(data, window=5):
    data_stretched = ([data[0]] * window) + data
    smoothed = []
    
    for i in xrange(0, len(data_stretched) - window):
        smoothed.append(np.average(data_stretched[i:i+window]))
        
    return smoothed

smoothed = smooth(frames_between_beats)
```

```python
setup_graph(x_label='Time', y_label='Heartrate (in bpm)')
plot([(float(video_fps) / delta) * 60 for delta in frames_between_beats], 'grey')
plot([(float(video_fps) / delta) * 60 for delta in smoothed])
```
