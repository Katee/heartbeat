```python
import subprocess as sp
import numpy as np

FFMPEG = "ffmpeg"
videofile = "heartbeat.mp4"

video_width = 1280
video_height = 720
```

```python
def get_red_from_frame(frame):
    """ Take a sampling from a frame and return the average amount of red """
    # points to sample red at, TODO sample randomly
    samples = [[500, 100], [200, 200], [600, 400]]
    
    reds = [frame[sample[0]][sample[1]][0] for sample in samples]
    return np.average(reds)
```
